```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("synthetic_microplastics_1s_200pts.csv")
first_row_arr = np.array(df.iloc[8,:200])
```

```python
print(first_row_arr)
```

```python
plt.style.use('seaborn-v0_8-darkgrid')

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 4))

# Plot the signal
ax.plot(first_row_arr, color='navy', linewidth=2, label='Signal')

# Add labels and title
ax.set_title('Input signal 1sec 200 readings', fontsize=14, fontweight='bold')
ax.set_xlabel('Time (samples)', fontsize=12)
ax.set_ylabel('Voltage', fontsize=12)

# Add grid and legend
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(loc='upper right', fontsize=10)

# Tight layout for clean spacing
plt.tight_layout()

# Show the plot
plt.show()
```
